```python
import matplotlib.pyplot as plt
import numpy as np

# Data for models
models = ['CNN', 'VGG16', 'ResNet50', 'DenseNet', 'Vision Transformer']
accuracy = [67.76, 72.83, 67.76, 78.40, 84.62]
sensitivity = [11.68, 19.29, 0.00, 39.59, 91.55]
specificity = [94.44, 98.31, 100.00, 96.86, 70.05]
f1_score = [0.19, 0.31, np.nan, 0.54, 0.89]

x = np.arange(len(models))

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar plot for Accuracy, Sensitivity, Specificity
bar_width = 0.2
ax1.bar(x - bar_width, accuracy, width=bar_width, label='Accuracy (%)')
ax1.bar(x, sensitivity, width=bar_width, label='Sensitivity (%)')
ax1.bar(x + bar_width, specificity, width=bar_width, label='Specificity (%)')

# Setup axes and title
ax1.set_xlabel('Models')
ax1.set_ylabel('Percentage')
ax1.set_title('Model Performance Comparison')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend()

# Line plot for F1-Score on the second y-axis
ax2 = ax1.twinx()
ax2.plot(x, f1_score, color='r', marker='o', label='F1-Score (Disease)')
ax2.set_ylabel('F1-Score')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
