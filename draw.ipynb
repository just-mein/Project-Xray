```python
import os

# Đường dẫn tới các thư mục dữ liệu
train_path = './train'
test_path = './test'

def count_images(data_path):
    categories = ['Disease', 'Normal']
    counts = {}
    for category in categories:
        category_path = os.path.join(data_path, category)
        if os.path.isdir(category_path):
            counts[category] = len(os.listdir(category_path))
    return counts

# Đếm số lượng hình ảnh trong tập huấn luyện và tập kiểm tra
train_counts = count_images(train_path)
test_counts = count_images(test_path)

print("Training data counts:", train_counts)
print("Testing data counts:", test_counts)
```

```python
import matplotlib.pyplot as plt

# Số lượng hình ảnh trong các tập dữ liệu
data_counts = {
    'train': {'Normal': 551, 'Disease': 1392},
    'test': {'Normal': 197, 'Disease': 414},
    'val': {'Normal': 145, 'Disease': 353}  # Thay đổi số lượng hình ảnh ở đây nếu cần thiết
}

categories = ['Normal', 'Disease']
datasets = ['train', 'test', 'val']
data = {dataset: [data_counts[dataset][category] for category in categories] for dataset in datasets}

# Vẽ biểu đồ
fig, ax = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
ax[0].set_ylim(0, 1600)
for i, dataset in enumerate(datasets):
    ax[i].bar(categories, data[dataset], color=['blue', 'orange'])
    ax[i].set_title(dataset.upper())
    ax[i].set_xlabel('Categories')
    ax[i].set_ylabel('Number of Images')
    for j, v in enumerate(data[dataset]):
        ax[i].text(j, v + 50, str(v), ha='center')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu cho các mô hình
models = ['CNN', 'VGG16', 'ResNet50', 'DenseNet121', 'Vision Transformer', 
          'EfficientNetB3', 'EfficientNetV2', 'DenseNet169', 'NASNet-A', 'ResNet101']

accuracy = [68.09, 70.54, 67.76, 79.05, 83.80, 
            79.00, 85.92, 86.74, 74.14, 67.76]

sensitivity = [96.14, 99.52, 100.00, 78.26, 81.88, 
               85.99, 83.09, 93.00, 99.03, 100.00]

specificity = [9.14, 9.64, 0.00, 80.71, 87.82, 
                64.47, 91.88, 73.60, 21.83, 0.00]

f1_score = [0.80, 0.82, 0.82, 0.84, 0.87, 
             0.85, 0.89, 0.90, 0.84, 0.81]

x = np.arange(len(models))

# Vẽ biểu đồ
fig, ax1 = plt.subplots(figsize=(14, 8))

# Vẽ biểu đồ cột cho Accuracy, Sensitivity, Specificity
bar_width = 0.2
ax1.bar(x - bar_width, accuracy, width=bar_width, label='Accuracy (%)')
ax1.bar(x, sensitivity, width=bar_width, label='Sensitivity (%)')
ax1.bar(x + bar_width, specificity, width=bar_width, label='Specificity (%)')

# Thiết lập trục và tiêu đề
ax1.set_xlabel('Models')
ax1.set_ylabel('Percentage')
ax1.set_title('Model Performance Comparison')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend()

# Vẽ biểu đồ đường cho F1-Score trên trục thứ hai
ax2 = ax1.twinx()
ax2.plot(x, f1_score, color='r', marker='o', label='F1-Score (Disease)')
ax2.set_ylabel('F1-Score')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
